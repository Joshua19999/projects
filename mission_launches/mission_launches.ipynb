```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df_mission_launches=pd.read_csv("mission_launches.csv", sep=",")
df_mission_launches = df_mission_launches.drop( 
    df_mission_launches.iloc[:,0:1].columns,
    axis=1
)
df_mission_launches
```

```python
meses=df_mission_launches.Date.str.split(pat="," and " ",expand=True).drop([0,2,3,4,5],axis=1)
anio=df_mission_launches.Date.str.split(pat="," and " ",expand=True).drop([0,1,2,4,5],axis=1)
df_mission_launches["Anio"]=anio.astype(int)
df_mission_launches["Meses"]=meses.astype(str)
df_mission_launches
```

```python
aniobus=2020
df_filtrado=df_mission_launches[["Organisation","Anio"]][df_mission_launches["Anio"]==aniobus]
organizaciones=df_filtrado.Organisation.unique()
Mission_Status=df_filtrado.Organisation.value_counts()

type(Mission_Status)
```

```python

plt.figure(figsize=(16, 6))
plt.bar(organizaciones, Mission_Status,color="r")
ax = plt.subplot()
plt.title("Orgaciones con mÃ¡s misiones")
ax.set_xlabel('Organizaciones')  # Nombre del eje x
ax.set_ylabel('Misiones')  # Nombre del eje y

```

```python
costo=df_mission_launches.Price
tiempo=df_mission_launches.Anio
costocada=[]
tiempocada=[]
acceder=0
cada=10
while acceder<len(costo):
    # print(costo[acceder],"--->",tiempo[acceder])
    costocada.append(costo[acceder])
    tiempocada.append(tiempo[acceder])
    acceder+=cada

costo=pd.Series(costocada)
tiempo=pd.Series(tiempocada)

df_precios=pd.DataFrame()
df_precios["Precios"]=costo
df_precios["Anios"]=tiempo

df_precios=df_precios.dropna()
costo=df_precios["Precios"]
tiempo=df_precios["Anios"]
costo,tiempo
```

```python
plt.figure(figsize=(18,12))
plt.bar(tiempo,costo,color="r")
ax = plt.subplot()
plt.title(f"Precios cada {cada} aÃ±os")
ax.set_xlabel('AÃ±os')  # Nombre del eje x
ax.set_ylabel('Precios')  # Nombre del eje y
```
